```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df=pd.read_csv('C:\\Users\\User\\Downloads\\android-games.csv')
df
```

```python
df['category'].value_counts()
```

```python
for cat in df.category:
    print(cat.split(' ')[1])
```

```python
text=" ".join(cat.split()[1] for cat in df.category)
text
```

```python
from wordcloud import WordCloud
word_cloud=WordCloud(height=2000,width=2000,background_color='black')
word_cloud=word_cloud.generate(text)
plt.imshow(word_cloud)
plt.title('most common words in positive customer comments')
plt.axis('off')
plt.show()
```
